# Data Visualization

## Matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
# univarite
#histograms
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
axes1.hist(tips['total_bill'], bins=10)
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('total bill')
fig.show
```

```python
#bivariate
#scatter plot - used when continuous variable vs continuous variable
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_title('Scatter plot of total bill vs tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
scatter_plot.show
```

```python
#box plot - discrete variabe vs continuous variable
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)
axes1.boxplot( 
    [tips[tips['sex'] == 'Female'] ['tip'], 
    tips[tips['sex'] == 'Male'] ['tip']],
    labels = ['Female', 'Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of tips by sex')
boxplot.show()
```

### Multivariate 

```python
# creating a colour variable based on sex
def recode_sex(sex):
    if sex =='Female':
        return 0
    else:
        return 1
tips['sex_colour'] = tips['sex'].apply(recode_sex)

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)

axes1.scatter(x = tips['total_bill'],      #
              y = tips['tip'],
              s=tips['size']*10, 
              c=tips['sex_colour'], alpha = 0.5)
            
axes1.set_title('Total bill vs Tip coloured by sex and sized by size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
scatter_plot.show
```

## Seaborn

```python
import seaborn as sns
```

```python
#univariate
#histograms
#histograms = distplot
hist,ax = plt.subplots()
ax = sns.distplot(tips['total_bill']) # #kde = False if you dont want the distplot line
ax.set_title('Total bill histogram with density plot')
plt.show()
```

```python
# density plot (kernel density estimation)
#created by drawing a normal distribution centered at each datapoint and smoothing out the overlapping plots so 
#that the area under the curve is 1

den,ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist= False)
axes1.set_title('Total bill Density')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Unit Probability')
plt.show()
```

```python
## Rug Plot
#one dimensional representation of a variable's distribution
#used with other plots to enhance visualization

# hist with density plot and rug plot on the bottom
hist_den_rug,ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = True) # #kde = False if you dont want the distplot line
ax.set_title('Total bill histogram with density and rug plot')
plt.show()
```

```python
#count plot - count discrete variables
count,ax = plt.subplots()
ax = sns.countplot('day', data=tips)
axes1.set_title('Count of days')
axes1.set_xlabel('Day of the week')
axes1.set_ylabel('Frequency')
plt.show()
```

## BiVariate

```python
#scatterplot
#scatter plot = regplot and a regression line

scatter,ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
axes1.set_title('Scatter plot of total bill vs tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
plt.show
```

```python
# hexbin plot 
#binning and scatter plots 
hexbin = sns.jointplot(x="total_bill", y = "tip", data=tips, kind="hex")
hexbin.set_axis_labels(xlabel='Total Bill', ylabel = 'Tip')
hexbin.fig.suptitle('Hexbin joint plot of total bill and tip', fontsize=10,y=1.03)
```

```python
## 2d density plot
kde,ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'], data2=tips['tip'],shade=True)

ax.set_title('Kernel Density plot of Total bill and tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
plt.show()

kde_joint = sns.jointplot(x='total_bill', y = 'tip', data=tips,kind='kde')
```

```python
# box plot 
box,ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Box plot of total bill by the time of the day')
axes1.set_xlabel('time of the day')
axes1.set_ylabel('Total bill')
plt.show()
```

```python
## Violin PLot
# this plot is able to show same values as a box plot but the boxes as kernel density estimation
# so you can retain more visual information of your data
violon,ax = plt.subplots()
ax = sns.violinplot(x='time', y = 'total_bill', data = tips)
ax.set_title('Violin plot of total bill by the time of the day')
axes1.set_xlabel('time of the day')
axes1.set_ylabel('Total bill')
plt.show()
```
